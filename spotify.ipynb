```python
import pandas as pd
import plotly.graph_objects as go


years = sorted(df['year'].unique())
fig = go.Figure()

# Create a trace for each year
for i, year in enumerate(years):
    top_artists = df[df["year"] == year]["artist_name"].value_counts().nlargest(10)

    fig.add_trace(go.Bar(
        x=top_artists.index,
        y=top_artists.values,
        name=str(year),
        visible=(i == 0)  # Show only the first year initially
    ))

# Create buttons for dropdown
buttons = [
    dict(label=str(year),
         method="update",
         args=[{"visible": [j == i for j in range(len(years))]},
               {"title": f"Top Artists in {year}"}])
    for i, year in enumerate(years)
]

# Update layout with dropdown
fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=0.1,
        y=1.2
    )],
    title="Top Artists by Year",
    xaxis_title="Artist Name",
    yaxis_title="Count",
    showlegend=False
)

fig.show(renderer='iframe')
```

```python
import pandas as pd
import plotly.graph_objects as go


years = sorted(df['year'].unique())
fig = go.Figure()

# Create a trace for each year
for i, year in enumerate(years):
    top_artists = df[df["year"] == year]["hour"].value_counts().nlargest(10)

    fig.add_trace(go.Bar(
        x=top_artists.index,
        y=top_artists.values,
        name=str(year),
        visible=(i == 0)  # Show only the first year initially
    ))

# Create buttons for dropdown
buttons = [
    dict(label=str(year),
         method="update",
         args=[{"visible": [j == i for j in range(len(years))]},
               {"title": f"Listen time in {year}"}])
    for i, year in enumerate(years)
]

# Update layout with dropdown
fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=0.1,
        y=1.2
    )],
    title="listen time by Year",
    xaxis_title="Hours",
    yaxis_title="Count",
    showlegend=False
)

fig.show(renderer='iframe')
```
