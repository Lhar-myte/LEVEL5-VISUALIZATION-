```python


# 1️⃣ Standardize column names
df3.columns = df3.columns.str.strip().str.lower().str.replace(" ", "_")

# 2️⃣ Remove duplicates
df3.drop_duplicates(inplace=True)

# 3️⃣ Convert numeric columns to proper types
numeric_cols = ['difficulty', 'reward', 'duration']
for col in numeric_cols:
    if col in df3.columns:
        df3[col] = pd.to_numeric(df3[col], errors='coerce')

# 4️⃣ Safely parse 'channels' column
def safe_parse_channels(x):
    if isinstance(x, str) and x.startswith('['):  # list-like string
        try:
            return ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):  # already a list
        return x
    else:
        return []  # handle NaN or anything unexpected

if 'channels' in df3.columns:
    df3['channels'] = df3['channels'].apply(safe_parse_channels)

# 5️⃣ Find all unique channels
unique_channels = sorted({ch for sublist in df3['channels'] if isinstance(sublist, list) for ch in sublist})

# 6️⃣ Create Yes/No indicator columns for each channel
for ch in unique_channels:
    df3[ch] = df3['channels'].apply(lambda x: 'Yes' if ch in x else 'No')
df3['channels'] = df3['channels'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
text_cols = ['offer_id', 'offer_type']
for col in text_cols:
    if col in df3.columns:
        df3[col] = df3[col].astype(str).str.strip().str.lower()


cleaned_path = "cleaned_offers.csv"
df3.to_csv(cleaned_path, index=False)


```

```python
import pandas as pd
import ast

# --- 1. Load and Clean Column Names ---
df2 = pd.read_csv('events.csv')

# Clean column names (lowercase, replace spaces with underscores)
df2.columns = df2.columns.str.strip().str.lower().str.replace(" ", "_")


# --- 2. Safely Convert 'value' Column to Dictionary ---

def safe_to_dict(x):
    """Safely converts a stringified Python literal (dict) to a dict object."""
    if isinstance(x, str) and x.startswith("{"):
        try:
            # ast.literal_eval is used to safely parse the dictionary string
            return ast.literal_eval(x)
        except:
            return {}
    return {}

# Apply the conversion function
df2['value'] = df2['value'].apply(safe_to_dict)

# --- 3. Extract Key/Value Pairs into New Columns ---
# The two most important key/value pairs are 'offer id' and 'amount'.

# A. Extract 'offer_id' (handles both 'offer id' and 'offer_id' keys)
df2['offer_id'] = df2['value'].apply(lambda d: d.get('offer id') or d.get('offer_id'))

# B. Extract 'amount'
df2['amount'] = df2['value'].apply(lambda d: d.get('amount'))

# C. Convert amount to numeric for filtering/calculations
df2['amount'] = pd.to_numeric(df2['amount'], errors='coerce')


# --- 4. Save the Result ---
df2.to_csv('cleaned_events_finally.csv', index=False)
```

```python
import pandas as pd

# --- Load all datasets ---
customers_df = pd.read_csv("cleaned_customer.csv")
events_df = pd.read_csv("cleaned_events_finally.csv")
offers_df = pd.read_csv("cleaned_offers.csv")

# --- Merge datasets ---
# 1️⃣ Merge offers → events (on offer_id)
merged_df = pd.merge(events_df, offers_df, on="offer_id", how="left")

# 2️⃣ Merge the result → customers (on customer_id)
merged_df = pd.merge(merged_df, customers_df, on="customer_id", how="left")

# --- Optional Cleaning ---
# Replace missing numeric values with 0 and fill blanks for clarity
numeric_cols = merged_df.select_dtypes(include="number").columns
merged_df[numeric_cols] = merged_df[numeric_cols].fillna(0)
merged_df = merged_df.fillna("Unknown")

# --- Save final dataset ---
merged_df.to_csv("merged_final.csv", index=False)

print("✅ Merge complete! File saved as 'merged_final.csv'")
print("Final Columns:")
print(merged_df.columns.tolist())
```
