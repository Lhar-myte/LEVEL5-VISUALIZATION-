```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Onyx Data - DataDNA Dataset Challenge - Mobile Phone Sales Dataset - May 2025.xlsx")
df.head()
```

```python
# Convert dates
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['End of Month'] = pd.to_datetime(df['End of Month'])

# Ensure numeric types
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Units_Sold'] = pd.to_numeric(df['Units_Sold'], errors='coerce')
df['Total_Revenue'] = pd.to_numeric(df['Total_Revenue'], errors='coerce')

# Checking for missing values
print(df.isnull().sum())
```

## 1. Top-Selling Mobile Brands

```python
brand_sales = df.groupby('Brand')['Units_Sold'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
brand_sales.plot(kind='bar', color='skyblue')
plt.title("Total Units Sold by Brand")
plt.xlabel("Brand")
plt.ylabel("Units Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 1b. Top Mobile Models by Units Sold

```python
model_sales = df.groupby('Mobile_Model')['Units_Sold'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
model_sales.plot(kind='bar', color='orange')
plt.title("Top 10 Best-Selling Models")
plt.xlabel("Mobile Model")
plt.ylabel("Units Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 1c. Revenue by Country

```python
country_revenue = df.groupby('Country')['Total_Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
country_revenue.plot(kind='bar', color='green')
plt.title("Total Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2 Sales by Storage Size

```python
storage_sales = df.groupby('Storage_Size')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,4))
storage_sales.plot(kind='barh', color='purple')
plt.title("Units Sold by Storage Size")
plt.xlabel("Units Sold")
plt.ylabel("Storage Size")
plt.tight_layout()
plt.show()
```

## 2b. Sales by Color

```python
color_sales = df.groupby('Color')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,4))
color_sales.plot(kind='barh', color='teal')
plt.title("Units Sold by Color")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
size_sale = df.groupby('Operating_System')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(8,5))
size_sale.plot(kind='bar', color='red')
plt.title("Units Sold by Operating ystem")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
age_group = df.groupby('Customer_Age_Group')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,3))
age_group.plot(kind='barh', color='green')
plt.title("Units Sold by customers age group")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
customer_gender = df.groupby('Customer_Gender')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,3))
customer_gender.plot(kind='bar', color='yellow')
plt.title("Units Sold by customers gender")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
sales_reveneu= df.groupby('Total_Revenue')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(11,5))
sales_reveneu.plot(x='Total_Revenue', y='Units_Sold', kind='line', marker='o', title='Units Sold by sales reveneu')
plt.title("Units Sold by sales reveneu")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
sales_channel= df.groupby('Sales_Channel')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,3))
sales_channel.plot(kind='barh', color='grey')
plt.title("Units Sold by sales channel")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
payment_types= df.groupby('Payment_Type')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,3))
payment_types.plot(kind='bar', color='black')
plt.title("Units Sold by payment type")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
price = df.groupby('Price')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(11,5))
price.plot(x='Price', y='Units_Sold', kind='line', marker='o', title='Units Sold by price')
plt.title("Units Sold by price")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
city_sales= df.groupby('City')['Units_Sold'].sum().sort_values()

plt.figure(figsize=(6,11))
city_sales.plot(kind='barh', color='grey')
plt.title("Units Sold by sales cities")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()
```
