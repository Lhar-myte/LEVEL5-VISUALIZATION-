```python
df["ts"] = pd.to_datetime(df["ts"])  #we converted the ts datatype to time object
#we fill all the missing values with blank
df["reason_start"] = df["reason_start"].fillna("Blank") 
df["reason_end"] = df["reason_start"].fillna("Blank")
```

```python
# working with date and time
df["year"] = df["ts"].dt.year
df["month"] = df["ts"].dt.month
df["day"] = df["ts"].dt.day
df["hour"] = df["ts"].dt.hour
df["minute"]= df["ts"].dt.minute
df["second"] = df["ts"].dt.second
```

```python
df.head()
```

```python
df["row"] = np.arange(1, 149861)
```

```python
df.head()
```

```python
df["year"].value_counts()
```

## LINE CHART

```python
day = df.groupby("day")["day"].count()
fig = px.line(data_frame=day)
fig.show(renderer='iframe')
```

# BAR CHART

```python
year = df.groupby("year")["day"].count()
fig = px.bar(data_frame=year,  title="YEAR WITH NUMBERS OF STREAM")
fig.show(renderer='iframe')
```

```python
year_month = df.groupby(["year", "month"])["day"].count().unstack().fillna(0)
```

```python
year_month
```

```python
fig = px.bar(data_frame=year_month,  title="YEAR BY MONTH WITH NUMBERS OF STREAM")
fig.show(renderer='iframe')
```

```python
year_skip = df.groupby(["year","skipped"])["day"].count().unstack().fillna(0)
```

```python
year_skip
```

```python
fig = px.bar(data_frame=year_skip, barmode="group")
fig.show(renderer='iframe')
```
